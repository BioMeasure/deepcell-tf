```python
import numpy as np
import skimage.io
import skimage.measure
import skimage.morphology
import matplotlib.pyplot as plt
```

```python
type_anno = skimage.io.imread('/data/data/tissues/mibi/mibi_full/set1/celltype/feature_0.tif')
type_anno_copy = np.copy(type_anno)
```

```python
plt.rcParams['figure.figsize'] = [10,10]
plt.imshow(type_anno, cmap='gnuplot2')
```

```python
eroded_binary = skimage.morphology.binary_erosion(type_anno)
eroded_type = eroded_binary*type_anno
plt.imshow(eroded_type, cmap='gnuplot2')
```

```python
dilated_binary = skimage.morphology.binary_dilation(type_anno)
dilated_type = dilated_binary*type_anno
plt.imshow(dilated_binary)
```

```python
print(dilated_binary.max())
```

```python
print(eroded_binary.min())
```

```python
pixels_to_rem = np.logical_and(dilated_binary, eroded_binary==False)
```

```python
plt.imshow(pixels_to_rem)
```

```python
type_anno_copy[pixels_to_rem] = type_anno_copy.max()+1
```
